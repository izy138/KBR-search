```python
from pathlib import Path
import pandas as pd



DATA_PATH = Path("data/ActivityCodes.csv")
if not DATA_PATH.exists():
    raise FileNotFoundError(f"Could not find data file at: {DATA_PATH.resolve()}")

df = pd.read_csv(DATA_PATH)

print(f"Loaded {len(df)} rows and {len(df.columns)} columns from {DATA_PATH}")
df.head()
```

```python
keep_cols = [
    "Activity Code",
    "Title" ]

df = df[keep_cols]

df.head()
```

```python
output_path = Path("../data")
df.to_csv(output_path, index=False) 

print(f"Saved CSV: {output_path.resolve()}")
print(f"Rows: {len(df):,}")
```
